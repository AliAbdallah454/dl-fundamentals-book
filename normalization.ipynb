```python
import torch
import torch.nn as nn
```

```python
layer1 = torch.tensor([1, 2, 3, 2], dtype=torch.float32).view(1, 4)
layer1.shape
```

```python
layer1
```

```python
mu = layer1.mean()
std = layer1.std(unbiased=False)
std_unbiased = layer1.std(unbiased=True)

mu, std, std_unbiased
```

```python
(layer1 - mu) / std
```

```python
img = torch.tensor([
    [0, 1, 0],
    [1, 1, 1],
    [0, 1, 0],
], dtype=torch.float32)
img.shape

img = torch.rand(3, 3, 4, dtype=torch.float32)
img.shape
```

```python
img_mu = img.mean()
img_std = img.std(unbiased=False)

print(img_mu, img_std)

(img - img_mu) / img_std
```

```python
layer_norm: torch.nn.LayerNorm = torch.nn.LayerNorm((3, 3, 4), eps=0)

layer_norm(img)
layer_norm.weight.shape, layer_norm.bias.shape
# layer_norm.weight
```
