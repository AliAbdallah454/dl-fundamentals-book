```python
import torch

import torch.optim as optim
```

```python
# MNIST Training, Conv -> Batch -> Activation -> Pooling
import torch.nn.functional as F
import torch.nn as nn

class MNIST_Net(nn.Module):

    def __init__(self):
        super().__init__()

        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, padding=1)
        self.batch_norm1 = nn.BatchNorm2d(16)
        
        self.conv2 = nn.Conv2d(16, 8, kernel_size=3, padding=1)
        self.batch_norm2 = nn.BatchNorm2d(8)

        self.fc1 = nn.Linear(8 * 7 * 7, 10)

    def forward(self, x):

        x = self.conv1(x)
        x = F.max_pool2d(F.relu(self.batch_norm1(x)), 2)

        x = self.conv2(x)
        x = F.max_pool2d(F.relu(self.batch_norm2(x)), 2)

        x = torch.flatten(x, 1, -1)
        x = self.fc1(x)

        return x
    
model = MNIST_Net()
```

```python
model = MNIST_Net()

optim.Adagrad(model.parameters(), lr=1e-3, initial_accumulator_value=0)

optim.RMSprop(model.parameters(), lr=1e-3, alpha=0.9)
```
